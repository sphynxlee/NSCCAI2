Create some toy data to work with. You will later - as an activity - experiment with some real data.

```python
import numpy as np

def generateFakeData(n,k):
  np.random.seed(10)
  points_per_cluster = float(n)/k
  X = []
  for i in range(k):
    stepsTakenCentroid = np.random.uniform(25000.0,3000.0)
    ageCentroid  = np.random.uniform(70.0,20.0)
    for j in range(int(points_per_cluster)):
      # this paper for stddev: https://www.ncbi.nlm.nih.gov/pmc/articles/PMC5774986/
      X.append([np.random.normal(stepsTakenCentroid,1500.0),np.random.normal(ageCentroid,2.0)])
  X = np.array(X)
  return(X)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs

# Set random seed for reproducibility
np.random.seed(42)

# Create synthetic data with 5 clusters
SynData, y = make_blobs(n_samples=500, centers=5, random_state=42, cluster_std=1.0)

# Visualize the generated data
plt.scatter(SynData[:, 0], SynData[:, 1], c=y, cmap='viridis', edgecolors='k')
plt.title("Synthetic Data with 5 Clusters")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.show()
```

```python
data = generateFakeData(100,5)
print(data)
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(8,6))
plt.scatter(SynData[:,0],SynData[:,1])
```

```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import scale


model = KMeans(n_clusters=4)
model = model.fit(scale(SynData))
print(model.labels_)

plt.figure(figsize=(8,6))
plt.scatter(SynData[:,0],SynData[:,1],c=model.labels_.astype(np.float64))
plt.show()
```
